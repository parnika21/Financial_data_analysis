## Data

We need to get data using pandas datareader. We will get stock information for the following banks:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo


```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
# Bank of America
BAC = data.DataReader("BAC", 'yahoo', start, end)

# CitiGroup
C = data.DataReader("C", 'yahoo', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

```python
BAC
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

** Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC],axis=1,keys=tickers)#concatenating the bank dataframes according to the tickers
bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

Now, checking the head of the bank_stocks dataframe.

```python
bank_stocks.head()
```

# Exploring the newly created DataSet 


Max Close price for each bank's stock.


```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

**Creating a new empty DataFrame called returns which will contain the returns for each bank's stock.Formula for returns:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns=pd.DataFrame()
```

** Using pct_change() method on the Close column to create a column representing this return value.**

```python
for i in tickers:
    returns[i+'Return']=bank_stocks[i]['Close'].pct_change()
    
returns.head()
```

** Creating a pairplot using seaborn of the returns dataframe.**

```python
import seaborn as sns
sns.pairplot(returns[1:])
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

Plotting the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008.

```python
plt.figure(figsize=(12,6))
BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 Day Avg')
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC CLOSE')
plt.legend()
```

Creating a heatmap of the correlation between the stocks Close Price.

```python
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```

clustering the correlations together

```python
sns.clustermap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```
